### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
df.columns.tolist()
df.dtypes
df.info()
df.describe()
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
missing_counts = df.isnull().sum()
missing_counts = missing_counts[missing_counts > 0].sort_values(ascending=False)
# print(missing_counts)

# For 'ConvertedCompYearly' (numerical, high missingness)
median_comp = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'].fillna(median_comp, inplace=True)

# For 'JobSat' (categorical or ordinal, high missingness)
mode_jobsat = df['JobSat'].mode()[0]
df['JobSat'].fillna(mode_jobsat, inplace=True)

# For JobSatPoints_1 to JobSatPoints_5 (numerical/ordinal)
for col in ['JobSatPoints_1', 'JobSatPoints_4', 'JobSatPoints_5']:
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# Value counts for Employment
employment_counts = df['Employment'].value_counts(dropna=False)
print(employment_counts)

# Value counts for JobSat (Job Satisfaction)
jobsat_counts = df['JobSat'].value_counts(dropna=False)
print(jobsat_counts)

# Value counts for YearsCodePro (Professional Coding Experience)
yearscodepro_counts = df['YearsCodePro'].value_counts(dropna=False)
print(yearscodepro_counts)
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here

# Prepare data (drop missing for clearer chart)
jobsat_counts = df['JobSat'].value_counts()

plt.figure(figsize=(8, 8))
jobsat_counts.plot.pie(autopct='%1.1f%%', startangle=140, cmap='Pastel1', ylabel='')

plt.title('Job Satisfaction Distribution (JobSat)')
plt.axis('equal')  # Equal aspect ratio for a perfect circle
plt.show()

# KDE
plt.figure(figsize=(8, 5))
sns.kdeplot(df['JobSat'].dropna(), shade=True, color='blue')

plt.title('Kernel Density Estimate of Job Satisfaction (JobSat)')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Density')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
# !pip install matplotlib-venn

from matplotlib_venn import venn2

# Drop rows with missing language data
have_df = df.dropna(subset=['LanguageHaveWorkedWith'])
want_df = df.dropna(subset=['LanguageWantToWorkWith'])

# Split languages into lists and strip whitespace
have_langs = have_df['LanguageHaveWorkedWith'].str.split(';').explode().str.strip()
want_langs = want_df['LanguageWantToWorkWith'].str.split(';').explode().str.strip()

# Count frequencies
have_counts = have_langs.value_counts()
want_counts = want_langs.value_counts()

# Create sets for overlap
have_set = set(have_counts.index)
want_set = set(want_counts.index)

# Print info about overlap
print(f"Unique languages in 'Have Worked With': {len(have_set)}")
print(f"Unique languages in 'Want To Work With': {len(want_set)}")
print(f"Languages in both sets (overlap): {len(have_set.intersection(want_set))}")
print(f"Example common languages: {list(have_set.intersection(want_set))[:10]}")

# Plot Venn diagram
plt.figure(figsize=(8,6))
venn2([have_set, want_set], set_labels=('Have Worked With', 'Want To Work With'))
plt.title('Overlap of Programming Languages')
plt.show()

#grouped bar chart

# Combine counts for top 10 languages overall
top_langs = pd.Index(have_counts.index).union(want_counts.index).tolist()
top_langs = have_counts.add(want_counts, fill_value=0).sort_values(ascending=False).head(10).index

# Create DataFrame for plotting
freq_df = pd.DataFrame({
    'Have Worked With': have_counts.reindex(top_langs).fillna(0),
    'Want To Work With': want_counts.reindex(top_langs).fillna(0)
})

freq_df.plot(kind='bar', figsize=(12,6), colormap='tab20')

plt.title('Comparison of Programming Languages: Have Worked With vs Want To Work With')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
# Drop missing values to simplify the visualization
remote_df = df.dropna(subset=['Country', 'RemoteWork'])

# Create a frequency table
remote_by_country = pd.crosstab(remote_df['Country'], remote_df['RemoteWork'])

# Optionally, keep only the top N countries by response count for readability
top_countries = remote_df['Country'].value_counts().head(10).index
remote_by_country = remote_by_country.loc[top_countries]

# grouped bar chart
remote_by_country.plot(kind='bar', figsize=(14, 6))

plt.title('Remote Work Distribution by Country')
plt.ylabel('Number of Respondents')
plt.xlabel('Country')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Remote Work Status')
plt.tight_layout()
plt.show()

# heat map
plt.figure(figsize=(12, 6))
sns.heatmap(remote_by_country, annot=True, fmt='d', cmap='YlGnBu')

plt.title('Heatmap of Remote Work Preferences by Country')
plt.xlabel('Remote Work Status')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
from scipy.stats import pearsonr, spearmanr

# Select and drop rows with missing values
df_corr = df[['JobSat', 'YearsCodePro']].dropna()

# Convert both columns from string to numeric
df_corr['JobSat'] = pd.to_numeric(df_corr['JobSat'], errors='coerce')
df_corr['YearsCodePro'] = pd.to_numeric(df_corr['YearsCodePro'], errors='coerce')

# Drop rows that failed conversion
df_corr = df_corr.dropna()

# Pearson (linear relationship)
pearson_corr, pearson_p = pearsonr(df_corr['YearsCodePro'], df_corr['JobSat'])

# Spearman (rank-based, monotonic relationship)
spearman_corr, spearman_p = spearmanr(df_corr['YearsCodePro'], df_corr['JobSat'])

print(f"📉 Pearson Correlation: {pearson_corr:.3f} (p = {pearson_p:.3f})")
print(f"📊 Spearman Correlation: {spearman_corr:.3f} (p = {spearman_p:.3f})")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
# Drop missing values to ensure a clean comparison
employment_edu = df.dropna(subset=['Employment', 'EdLevel'])

# Create a cross-tabulation (count of combinations)
crosstab = pd.crosstab(employment_edu['Employment'], employment_edu['EdLevel'])

# Plot stacked bar chart
crosstab.plot(kind='bar', stacked=True, figsize=(14, 6), colormap='tab20')

plt.title('Education Level Distribution by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
