### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
df['Employment'].value_counts(dropna=False)
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
df['EdLevel'].value_counts(dropna=False).index.tolist()

edlevel_mapping = {
    "Bachelor’s degree (B.A., B.S., B.Eng., etc.)": "Bachelor's",
    "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)": "Master's",
    "Some college/university study without earning a degree": "Some college",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "Secondary school",
    "Primary/elementary school": "Primary/elementary school",
    "Associate degree (A.A., A.S., etc.)": "Associate",
    "Professional degree (JD, MD, Ph.D, Ed.D, etc.)": "Doctorate",
    "Something else": 'Something else'
}

df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)

print(df['EdLevel'].value_counts().index.tolist())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
pd.get_dummies(df['Employment'], prefix='Employment')
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
# Calculate missing values per column
missing_counts = df.isnull().sum()

# Sort columns by missing count descending and show top columns
missing_sorted = missing_counts.sort_values(ascending=False)

print(missing_sorted[missing_sorted > 0])
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
# Impute with mean
mean_value = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly'].fillna(mean_value, inplace=True)

# OR, impute with median (recommended for skewed data)
median_value = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'].fillna(median_value, inplace=True)

# Verify no missing values remain
missing_after = df['ConvertedCompYearly'].isnull().sum()
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# Find the most frequent value in RemoteWork
most_frequent = df['RemoteWork'].mode()[0]

# Fill missing values with the mode
df['RemoteWork'].fillna(most_frequent, inplace=True)

# Verify no missing values remain
missing_after = df['RemoteWork'].isnull().sum()
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
df['ConvertedCompYearly_Normalized'] = ((df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) / (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min()))
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import numpy as np
df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
def categorize_experience(years):
    try:
        years = float(years)
    except:
        return "Unknown"  # for non-numeric or missing values
    
    if years < 1:
        return "Entry"
    elif 1 <= years < 3:
        return "Junior"
    elif 3 <= years < 5:
        return "Mid"
    elif 5 <= years < 10:
        return "Senior"
    else:
        return "Expert"

# Apply the function to create ExperienceLevel
df['ExperienceLevel'] = df['YearsCodePro'].apply(categorize_experience)
```
